```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading Dataset**

```python
df = pd.read_csv("/content/Food_Truck_Data.csv")
df.head(10)
```

```python
sns.scatterplot(x="Population", y="Profit", data=df)
plt.xlabel("Population")
plt.ylabel("Profit")
plt.title("Food Truck Profit vs Population")
plt.show()
```

```python
def cost_function(X,y,theta):
  x = len(y)
  y_pred = X.dot(theta)
  error = (y_pred - y) ** 2
  return 1/(2*m) * np.sum(error)
```

```python
m = df.Population.size
x = np.append(np.ones ((m,1)), df.Population.values.reshape(m,1), axis=1)
y = df.Profit.values.reshape (m,1)
theta = np.zeros ((2,1))

cost_function (x, y, theta)
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
  a= len(y)
  cost = []
  for i in range(iterations):
    y_pred = X.dot(theta)
    error = np.dot(X.transpose(), (y_pred - y))
    theta -= alpha * 1/m * error
    cost.append(cost_function(X, y, theta))
  return theta, cost
```

```python
theta, cost = gradient_descent(x, y, theta,alpha = 0.01, iterations = 10)
print("h(x) = {} + {}x1".format(str(round(theta[0,0],2)), str(round(theta[1,0],2))))
```
