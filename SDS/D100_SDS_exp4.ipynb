21. Verify Central Limit Theorem for Sampling distribution for Sample
mean for randomly generated non-normal population using Python.

```python
population = np.random.exponential(scale=1, size=10000)

sample_size = 30
num_samples = 1000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Original Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Sampling Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

stat, p_value = normaltest(sample_means)

print(f'Normality test statistic: {stat:.3f}, p-value: {p_value:.3f}')

if p_value > 0.05:
    print("Sample means follow a normal distribution (fail to reject null hypothesis).")
else:
    print("Sample means do not follow a normal distribution (reject null hypothesis).")
```

22. Verify Central Limit Theorem for Sampling distribution for Sample
mean for randomly generated normal population using Python.

```python
population = np.random.normal(scale=1, size=10000)

sample_size = 30
num_samples = 1000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(10, 3))

plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Original Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Sampling Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
stat, p_value = normaltest(sample_means)

print(f'Normality test statistic: {stat:.3f}, p-value: {p_value:.3f}')

if p_value > 0.05:
    print("Sample means follow a normal distribution (fail to reject null hypothesis).")
else:
    print("Sample means do not follow a normal distribution (reject null hypothesis).")
```

23. We have population data for individual smoking habits. We know
that the true population proportion for smoking is 0.395 or 39.5%. For
the given data verify Central Limit Theorem for Sampling distribution
for Sample proportions using Python. Take population size equal to
1000

```python
population_size = 1000
population_proportion = 0.395
sample_size = 50
num_samples = 1000

population = np.random.choice([0, 1], size=population_size, p=[1-population_proportion, population_proportion])

sample_proportions = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_proportions.append(np.mean(sample))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.hist(population, bins=2, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Population Distribution (Smokers vs Non-Smokers)')
plt.xlabel('Smokers (1) vs Non-Smokers (0)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sample_proportions, bins=30, color='orange', edgecolor='black', alpha=0.7)
plt.title('Sampling Distribution of Sample Proportions')
plt.xlabel('Sample Proportion')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

stat, p_value = normaltest(sample_proportions)

print(f'Normality test statistic: {stat:.3f}, p-value: {p_value:.3f}')

if p_value > 0.05:
    print("Sample proportions follow a normal distribution (fail to reject null hypothesis).")
else:
    print("Sample proportions do not follow a normal distribution (reject null hypothesis).")
```
