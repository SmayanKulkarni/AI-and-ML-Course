```python
p = 10000000
r = 9/(12*100)
n = 20*12
```

```python
def emi(p, r, n):
  num = p * r * ((1+r)**n)
  den = ((1+r)**n)-1
  return int(num/den)

e = emi(p, r, n)
print(e)
```

```python
def mi(p, r):
  mi = p * r
  return int(mi)
```

```python
def pp(p, r, n):
  pp = emi(p, r, n) - mi(p, r)
  return int(pp)
```

```python
balance = p
data = []
for i in range(1, n+1):
  interest= mi(balance, r)
  principal = pp(balance, r, n - i + 1)
  balance -= principal
  data.append([i, e, interest, principal, int(balance)])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data, columns = ['Month', 'EMI', 'Interest', 'Principal', 'Balance'])
df.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.stackplot(df['Month'], df['Interest'], df['Principal'], labels=['Interest', 'Principal'])
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Principal and Interest over Time')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Balance'])
plt.xlabel('Month')
plt.ylabel('Loan Balance')
plt.title('Remaining Loan Balance over Time')
plt.grid(True)
plt.show()
```
