```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock_price = 50 # Stock price at the time of option purchase
strike_price = 50 # Strike price for Long Call and Long Put
strike_price_cc = 60 # Strike price for Covered Call
strike_price_call = 50 # Strike price for Bull Call Spread (Long)
strike_price_short_call = 60 # Strike price for Bull Call Spread (Short)
option_cost = 5 # Cost of the call/put option (assumed to be $5 per option)
time_to_expiration = 30 / 365 # Time to expiration in years (30 days)
volatility = 0.20 # Volatility (20%)
risk_free_rate = 0.02 # Risk-free rate (2%)
S_range = np.linspace(40, 80, 200) # Stock Price Range for plotting payoff profiles
```

```python
long_call_payoff = np.maximum(S_range - strike_price, 0)
print("The long call payoffs between price range 150 - 200 are : \n", long_call_payoff[150:200])
```

```python
long_put_payoff = np.maximum(strike_price - S_range, 0)
print("The long put payoffs between price range 10 - 20 are : \n", long_put_payoff[10:20])
```

```python
covered_call_payoff = (S_range - stock_price) + np.maximum(S_range - strike_price_cc, 0)
print("The covered call payoffs between price range 10 - 20 are : \n", covered_call_payoff[10:20])
```

```python
bull_call_spread_payoff = np.maximum(S_range - strike_price_call, 0) - np.maximum(S_range - strike_price_short_call, 0)
print("The bull call spread payoffs between price range 100 - 130 are : \n", bull_call_spread_payoff[100:130])
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(S_range, long_call_payoff, label="Long Call", color="green",
lw=2)
plt.plot(S_range, long_put_payoff, label="Long Put", color="red", lw=2)
plt.plot(S_range, covered_call_payoff, label="Covered Call",
color="blue", lw=2)
plt.plot(S_range, bull_call_spread_payoff, label="Bull Call Spread",
color="purple", lw=2)
plt.title("Payoff Profiles for Various Option Strategies")
plt.xlabel("Stock Price at Expiration ($)")
plt.ylabel("Payoff ($)")
plt.axhline(0, color="black", lw=1)
plt.axvline(stock_price, color="gray", linestyle="--")
plt.axvline(strike_price, color="red", linestyle="--")
plt.axvline(strike_price_cc, color="blue", linestyle="--")
plt.axvline(strike_price_call, color="purple", linestyle="--")
plt.axvline(strike_price_short_call, color="purple", linestyle="--")
plt.legend(loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()
```
