```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amount = 500000
rate = 8/1200
n = 5*12
emi = (amount * rate * ((1+rate)**n)) / (((1+rate)**n) - 1)
print(round(emi,2))
```

```python
def calc_interest(amount=amount, rate=rate):
    interest = amount * rate
    return round(interest,3)
```

```python
interest = calc_interest()
```

```python
print(interest)
```

```python
def calc_principal(emi=emi, interest = interest):
    principal = emi - interest
    return round(principal,2)
```

```python
principal = calc_principal()
```

```python
principal
```

```python
newamount = amount
newrate = rate
price = []
princs = []
interest = []
```

```python
for i in range(1,n+1):
    newinterest = calc_interest(newamount , newrate)
    newprincipal = calc_principal(emi, newinterest)
    newamount = max(newamount - newprincipal,0)
    price.append(newamount)
    princs.append(newprincipal)
    interest.append(newinterest)
```

```python
import pandas as pd
```

```python
plt.title("Prices over months")
plt.xlabel("Months")
plt.plot(princs)
```

```python
plt.title("Remaining balance over the months")
plt.plot(price)
```

```python
df = pd.DataFrame()
df['Principal'] = np.array(princs)
df['Interest'] = np.array(interest)
```

```python

df['Month'] = np.arange(1,61)
df['Balance'] = np.array(price)
```

```python
plt.figure(figsize=(12, 6))
plt.stackplot(df['Month'], df['Interest'], df['Principal'], labels=['Interest', 'Principal'])
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Principal and Interest over Time')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Balance'])
plt.xlabel('Month')
plt.ylabel('Loan Balance')
plt.title('Remaining Loan Balance over Time')
plt.grid(True)
plt.show()
```
