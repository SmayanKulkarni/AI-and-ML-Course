## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#1. Prepare Data
x = [2,234,234,7,475,45]
y = [56,8,2,86,65,5,8]

#2. Setup Plot
fig,ax = plt.subplots(figsize=(5,5))

#3. Plot the data
ax.plot(x[:5],y[:5])

#4. Customize plot
ax.set(title='A Simple Plot', xlabel='X Axis', ylabel='Y Axis')

#5. Save and show
fig.savefig('/home/smayan/Desktop/AI-ML-DS/sample_project/images/sample_plot.png')
plt.show()
```
