## Two options for subplots

```python
#Subplot option 1

fig,((ax0,ax1), (ax2,ax3)) = plt.subplots(nrows = 2, ncols=2, figsize = (10,5))
ax0.plot(x, x*5+120)
ax0.set(title = 'A subplot', ylabel='Y axis', xlabel = 'X Axis')
ax1.hist(x)

ax2.scatter(x, np.random.randn(100))
ax3.barh(list(x+23*9),list( x+205))
```

```python
#Subplot option 2
fig,ax = plt.subplots(nrows = 2, ncols=2, figsize = (10,5))
ax[0,0].plot(x, x*5+120)
ax[0,0].set(title = 'A subplot', ylabel='Y axis', xlabel = 'X Axis')
ax[0,1].hist(x)

ax[1,0].scatter(x, np.random.randn(100))
ax[1,1].barh(list(x+23*9),list( x+205))
```

```python
heart_disease = pd.read_csv("/home/smayan/Desktop/AI-ML-DS/sample_project/heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease.plot.hist(subplots=True, bins=30,figsize = (10,25))
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind='scatter', x="age", y="chol", c='target');
```

```python
# OO method used with pyplot
fig,ax = plt.subplots(figsize=(10,6));
over_50.plot(kind='scatter',x="age", y="chol", c='target',ax=ax );
ax.set_xlim([50,75])
```

```python
over_50.target.values
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize = (10,5))

#Plotting the data
scatter =ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

#Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel ='age',
       ylabel='Cholesterol');

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

#Add a horizontal line
ax.axhline(over_50["chol"].mean(numeric_only = True),
           linestyle = "dashdot")
```

```python
over_50.head()
```

```python
#Subplot of chol, age and thalach
fig, (ax0,ax1) = plt.subplots(nrows =2,
                              ncols=1,
                              figsize=(15,10))
#Add data to ax0
scatter = ax0.scatter(x =over_50["age"], y=over_50["chol"], c = over_50["target"])
#customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol');
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(numeric_only = True),
           linestyle = "dashdot");

#Add data to ax1
scatter =ax1.scatter(x=over_50["age"], y = over_50["thalach"], c=over_50["target"])

#customize ax1
ax1.set(title="Heart Disease and Thalach",
        xlabel = "Age",
        ylabel = 'Max Heart Rate');
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(numeric_only=True),linestyle="dashdot")

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold")
```
