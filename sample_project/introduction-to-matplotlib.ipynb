## Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#1. Prepare Data
x = [2,234,234,7,475,45]
y = [56,8,2,86,65,5,8]

#2. Setup Plot
fig,ax = plt.subplots(figsize=(5,5))

#3. Plot the data
ax.plot(x[:5],y[:5])

#4. Customize plot
ax.set(title='A Simple Plot', xlabel='X Axis', ylabel='Y Axis')

#5. Save and show
fig.savefig('/home/smayan/Desktop/AI-ML-DS/sample_project/images/sample_plot.png')
plt.show()
```

## Making figures with NumPy arrays

```python
#Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2)
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.cos(x))
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.tan(x))
```

```python
butter_prices = {"Almond Butter" : 10, "Peanut Butter" : 20, "Cashew Butter" : 30}
fig,ax = plt.subplots(figsize = (5,4))
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title='Butters', xlabel='Flavors', ylabel='Price');
```
